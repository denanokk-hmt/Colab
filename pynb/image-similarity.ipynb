# [PythonとKerasでインスタの画像検索システム](https://medium.com/@toddperry_78831/pythonとkerasでインスタの画像検索システム-3b66f602badb)

```
def imageResize(img, h, w):
    ih, iw = img.shape[0], img.shape[1]
    if ih >= iw:
        dst = np.ones([ih, ih, 3], np.uint8) * 255
        dst[:, (ih - iw)//2:(ih - iw)//2 + iw] = img
    else:
        dst = np.ones([iw, iw, 3], np.uint8) * 255
        dst[(iw - ih)//2:(iw - ih)//2 + ih, :] = img
        
    dst = np.array(Image.fromarray(dst).resize((w, h)))
    return dst        
```

```
# KerasからVGG16(Imgnet)をインポートする
#from keras.applications.vgg16 import VGG16, preprocess_input
# 他のインポート
from keras.models import Model
from keras.preprocessing import image
from scipy import spatial
import numpy as np

# モデルを宣言する
#model = VGG16(weights='imagenet', include_top=False)

# 下の関数は画像のパスをもらって、一次元のべクターを返す
def do_feature_extraction(img_path):
    # パスで画像を読み込む
    #img = image.load_img(img_path, target_size=(224, 224))
    #img = image.load_img(img_path, target_size=(h, w))
    # 画像をピクセル配列に変更する
    #imgarr = image.img_to_array(img)
    img = imread(img_path)
    if img.shape[-1] == 4:
        img = img[:,:,0:3]
    imgarr = imageResize(img, h, w)
    imgarr = np.expand_dims(imgarr, axis=0)
    imgarr = preprocess_input(imgarr)
    # 処理したピクセル情報を深層学習モデルに与える
    #features = model.predict(imgarr)
    features = model_notop.predict(imgarr)
    # 返した三次元行列を一次元のべクターに変更する
    #feature_vector = features.reshape(7*7*512)    
    feature_vector = features.flatten()
    return feature_vector
  
# 下の関数は二つのべクターの相似を計算する
def get_vector_similarity(vec1, vec2):
    return 1 - spatial.distance.cosine(vec1, vec2)
  
# この関数は二つのJPGのパスをもらって、相似の値を返す
def get_image_similarity(path1, path2):
    # べクターを計算する
    vec1 = do_feature_extraction(path1)
    vec2 = do_feature_extraction(path2)
    # 相似を計算して返す
    return get_vector_similarity(vec1, vec2)
```

```
import numpy as np

d = 64                           # dimension
nb = 100000                      # database size
nq = 10000                       # nb of queries
np.random.seed(1234)             # make reproducible
xb = np.random.random((nb, d)).astype('float32')
xb[:, 0] += np.arange(nb) / 1000.
xq = np.random.random((nq, d)).astype('float32')
xq[:, 0] += np.arange(nq) / 1000.

import faiss                   # make faiss available
index = faiss.IndexFlatL2(d)   # build the index
print(index.is_trained)
index.add(xb)                  # add vectors to the index
print(index.ntotal)

k = 4                          # we want to see 4 nearest neighbors
D, I = index.search(xb[:5], k) # sanity check
print(I)
print(D)
D, I = index.search(xq, k)     # actual search
print(I[:5])                   # neighbors of the 5 first queries
print(I[-5:])                  # neighbors of the 5 last queries
```
