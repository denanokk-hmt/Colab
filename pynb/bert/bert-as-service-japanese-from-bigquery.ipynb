## [Building a QA semantic search engine in 3 minutes](https://bert-as-service.readthedocs.io/en/latest/tutorial/simple-search.html)

```
!wget https://raw.githubusercontent.com/lvisdd/qnabot/master/webapp/data/faq.tsv
```

```
from pyknp import Juman

def tokenize(str):
    jumanpp = Juman()
    result = jumanpp.analysis(str)
    midasi = [mrph.midasi for mrph in result.mrph_list()]
    return " ".join(midasi);
```

```
import csv
import numpy as np

questions=[]

with open('faq.tsv', encoding="utf_8") as file:
    reader = csv.reader(file, delimiter='\t')
    for row in reader:
        #print(row)
        #questions.append(row[0])
        text = tokenize(row[0])
        print(text)
        questions.append(text)

print('%d questions loaded, avg. len of %d' % (len(questions), np.mean([len(d.split()) for d in questions])))
```

```
bc = BertClient(port=5555, port_out=5556)
doc_vecs = bc.encode(questions)
```

```
topk = 5

print("ex)")
print("「みぞれ」とは何ですか？")
print("q (終了)")

while True:
    query = input('your question: ')

    if query == 'q':
        break
    
    #query_vec = bc.encode([query])[0]
    query_vec = bc.encode([tokenize(query)])[0]
    # compute normalized dot product as score
    score = np.sum(query_vec * doc_vecs, axis=1) / np.linalg.norm(doc_vecs, axis=1)
    topk_idx = np.argsort(score)[::-1][:topk]
    for idx in topk_idx:
        #print('> %s\t%s' % (score[idx], questions[idx]))
        print('> %s\t%s' % (score[idx], "".join(questions[idx].split())))
```
