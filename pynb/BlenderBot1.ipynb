```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration

# モデルとトークナイザーの準備
mname = 'facebook/blenderbot-400M-distill'
model = BlenderbotForConditionalGeneration.from_pretrained(mname)
tokenizer = BlenderbotTokenizer.from_pretrained(mname)
```

```python
# BlenderBotと対話
UTTERANCE = "My friends are cool but they eat too many carbs."
inputs = tokenizer([UTTERANCE], return_tensors='pt')
reply_ids = model.generate(**inputs)
print(tokenizer.batch_decode(reply_ids))
```

## 2. 様々な対話を試す

```python
from googletrans import Translator

def translate(text, src='en', dest="ja"):
    translator = Translator()
    translated = translator.translate(text, src='en', dest="ja");
    print(text) # English
    print(translated.text) # Japanese
```

```python
translate(UTTERANCE)
```

```python
import re

pattern = r"<[/ ]?s> ?"

def generate(text):
     inputs = tokenizer([text], return_tensors='pt')
     reply_ids = model.generate(**inputs)
     translate(text)
     result = re.sub(pattern, '', tokenizer.batch_decode(reply_ids)[0])
     # translate(tokenizer.batch_decode(reply_ids)[0])
     translate(result)
```

```python
generate("My friends are cool but they eat too many carbs.")
```

```python
generate("What kind of movie do you like.")
```

```python
generate("I like the movie Evangelion. did you see?")
```

```python
generate("It was an interesting movie. Let's go to the cinema together next time.")
```
